```python
import pandas as pd

merged_gw = pd.read_csv ('merged_gw.csv')
teams_fifa23 = pd.read_csv ('teams_fifa23.csv')
players_fifa23 = pd.read_csv ('players_fifa23.csv')
Odds = pd.read_csv ('E0.csv')
Team = pd.read_csv ('Teams.csv')
```

```python
ts = Team[['name', 'strength_overall_home', 'strength_overall_away', 
            'strength_attack_home', 'strength_attack_away',
            'strength_defence_home', 'strength_defence_away']]
team_strength = ts.transpose()
team_strength.columns = team_strength.iloc[0] 
team_strength = team_strength[1:]
```

```python
#ICT score dataset
merged_gw_updated = merged_gw[["team", "ict_index", "GW"]]
sorted_gw = merged_gw_updated.sort_values(by=['team', 'GW'])
grouped_gw = sorted_gw.groupby(by=['team', 'GW'])['ict_index'].sum()
df_grouped_gw = grouped_gw.to_frame()
ICT_2022 = df_grouped_gw.pivot_table(index=['team'], 
            columns=['GW'], values='ict_index').fillna(0)
ICT_2022.sort_index(inplace=True)
ICT_2022.reset_index(inplace=True)
```

```python
#was home team dataset
merged_gw_updated2 = merged_gw[["team", "was_home", "GW"]]
sorted_gw2 = merged_gw_updated2.sort_values(by=['team', 'GW'])
grouped_gw2 = sorted_gw2.groupby(by=['team', 'GW'])['was_home'].sum()
df_grouped_gw2 = grouped_gw2.to_frame()
WH_2022 = sorted_gw2.pivot_table(index=['team'], 
            columns=['GW'], values='was_home').fillna(0)
WH_2022.sort_index(inplace=True)
WH_2022.reset_index(inplace=True)
```

```python
away_team_ict = ICT_2022.mask(WH_2022 == 1.0)
away_team_ict = away_team_ict.rename({'team': 'Away Team', 1: 'GW1', 2: 'GW2', 3: 'GW3', 4: 'GW4', 
                            5: 'GW5', 6: 'GW6', 8: 'GW8', 9: 'GW9', 10: 'GW10'}, axis=1) 
```

```python
new_odds = Odds[["Date", "GW","HomeTeam", "AwayTeam", 
      "FTHG", "FTAG", "FTR", "Referee", 
      "HS", "AS", "HST", "AST", "HC", "AC", 
      "HY", "AY", "HR", "AR", "B365H", "B365D", "B365A"]]
new_odds = new_odds.assign(HICT="")
new_odds = new_odds.assign(AICT="")
new_odds.sort_index(inplace=True)
new_odds = new_odds.reset_index(drop=True)
new_odds = new_odds.sort_values('AwayTeam')
```

```python
gw_1 = new_odds[new_odds.GW == 1]
away_team_ict_g_1 = away_team_ict.GW1
away_team_ict_g_1 = away_team_ict_g_1.dropna()
away_team_ict_g_1 = away_team_ict_g_1.reset_index(drop=True)
gw_1 = gw_1.reset_index(drop=True)
gw_1 = gw_1.assign(AICT=away_team_ict_g_1)
```

```python
gw_2 = new_odds[new_odds.GW == 2]
away_team_ict_g_2 = away_team_ict.GW2
away_team_ict_g_2 = away_team_ict_g_2.dropna()
away_team_ict_g_2 = away_team_ict_g_2.reset_index(drop=True)
gw_2 = gw_2.reset_index(drop=True)
gw_2 = gw_2.assign(AICT=away_team_ict_g_2)
```

```python
hello = pd.concat([gw_1,gw_2], ignore_index=True)
hello
```

```python
hello['Date'] = pd.to_datetime(hello['Date'], infer_datetime_format=True)
final_dataset = hello.sort_values(by='Date').reset_index(drop=True)
```
